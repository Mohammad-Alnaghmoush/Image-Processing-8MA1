```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Setting the default figure size for better visibility in Jupyter
plt.rcParams['figure.figsize'] = [15, 5]
```

```python
#Task1
img1 = cv2.imread('image1.jpg') 
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

box_blur = cv2.blur(img1_rgb, (7, 7))

fig, ax = plt.subplots(1, 2)
ax[0].imshow(img1_rgb)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(box_blur)
ax[1].set_title("7x7 Box Filter (Smoothed)")
ax[1].axis("off")
plt.show()
```

```python
#Task2
img2 = cv2.imread('image2.jpg')
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)


gaussian_5x5 = cv2.GaussianBlur(img2_rgb, (5, 5), 0)
gaussian_21x21 = cv2.GaussianBlur(img2_rgb, (21, 21), 0)


fig, ax = plt.subplots(1, 3)
ax[0].imshow(img2_rgb)
ax[0].set_title("Original")
ax[0].axis("off")

ax[1].imshow(gaussian_5x5)
ax[1].set_title("5x5 Gaussian Blur")
ax[1].axis("off")

ax[2].imshow(gaussian_21x21)
ax[2].set_title("21x21 Gaussian Blur")
ax[2].axis("off")
plt.show()
```

```python
#Task3
img3 = cv2.imread('image3.jpg')
img3_rgb = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)


image_float = img3_rgb.astype(np.float32) / 255.0


laplacian = cv2.Laplacian(image_float, cv2.CV_32F)


sharpened = np.clip(image_float - laplacian, 0, 1)


fig, ax = plt.subplots(1, 3, figsize=(18, 6))

ax[0].imshow(image_float)
ax[0].set_title("Original Image")


ax[1].imshow(np.clip(np.abs(laplacian), 0, 1), cmap='gray')
ax[1].set_title("Laplacian (Edges)")

ax[2].imshow(sharpened)
ax[2].set_title("Sharpened Image (Clipped)")

plt.show()
```
