```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imgFile = "moon.jpg"
img = cv2.imread(imgFile, 0)

if img is None:
    print("Error: Image not found. Make sure moon.jpg is in the same folder.")
else:
    print("Image loaded successfully.")
    print("Image shape:", img.shape)
```

```python
#Mean Filter / Average Filter
mean_filter_3x3 = cv2.blur(img, (3, 3))
mean_filter_5x5 = cv2.blur(img, (5, 5))
```

```python
#Gaussian Filter
gaussian_filter_3x3 = cv2.GaussianBlur(img, (3, 3), 0)
gaussian_filter_5x5 = cv2.GaussianBlur(img, (5, 5), 0)
```

```python
#Median Filter
median_filter_3x3 = cv2.medianBlur(img, 3)
median_filter_5x5 = cv2.medianBlur(img, 5)
```

```python
#Min Filter
kernel = np.ones((3, 3), np.uint8)
min_filter = cv2.erode(img, kernel)
```

```python
#Max Filter
max_filter = cv2.dilate(img, kernel)
```

```python
#Display all results
plt.figure(figsize=(14, 10))

plt.subplot(3, 3, 1)
plt.title("Original Image")
plt.imshow(img, cmap="gray")
plt.axis("off")

plt.subplot(3, 3, 2)
plt.title("Mean Filter 3x3")
plt.imshow(mean_filter_3x3, cmap="gray")
plt.axis("off")

plt.subplot(3, 3, 3)
plt.title("Mean Filter 5x5")
plt.imshow(mean_filter_5x5, cmap="gray")
plt.axis("off")

plt.subplot(3, 3, 4)
plt.title("Gaussian Filter 3x3")
plt.imshow(gaussian_filter_3x3, cmap="gray")
plt.axis("off")

plt.subplot(3, 3, 5)
plt.title("Gaussian Filter 5x5")
plt.imshow(gaussian_filter_5x5, cmap="gray")
plt.axis("off")

plt.subplot(3, 3, 6)
plt.title("Median Filter 3x3")
plt.imshow(median_filter_3x3, cmap="gray")
plt.axis("off")

plt.subplot(3, 3, 7)
plt.title("Median Filter 5x5")
plt.imshow(median_filter_5x5, cmap="gray")
plt.axis("off")

plt.subplot(3, 3, 8)
plt.title("Min Filter")
plt.imshow(min_filter, cmap="gray")
plt.axis("off")

plt.subplot(3, 3, 9)
plt.title("Max Filter")
plt.imshow(max_filter, cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
#Short comparison
print("Mean Filter: Smooths the image by averaging neighboring pixels.")
print("Gaussian Filter: Smooths the image with more weight on the center pixel.")
print("Median Filter: Good for removing salt-and-pepper noise.")
print("Min Filter: Makes bright areas smaller and dark areas larger.")
print("Max Filter: Makes bright areas larger and dark areas smaller.")
```
