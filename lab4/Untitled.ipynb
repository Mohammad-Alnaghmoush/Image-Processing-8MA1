```python
#Task_1
import matplotlib.pyplot as plt
import numpy as np
from skimage import data, exposure

# Load the moon image from skimage data [cite: 47, 49, 50]
img = data.moon()

# Calculate the 3rd and 80th percentiles [cite: 53, 78]
p3, p80 = np.percentile(img, (3, 80))

# Rescale intensity values based on the calculated percentiles [cite: 54, 78]
img_rescale = exposure.rescale_intensity(img, in_range=(p3, p80))

# Display the rescaled image and its histogram [cite: 55, 78]
fig = plt.figure(figsize=(10, 8))


fig.add_subplot(2, 1, 1)
plt.imshow(img_rescale, cmap='gray')
plt.title('Moon Image - Rescaled (3rd to 80th Percentile)')
plt.axis('off')

# Subplot for the histogram [cite: 62, 78]
fig.add_subplot(2, 1, 2)
plt.hist(img_rescale.flat, bins=256, range=(0, 255))
plt.title('Histogram of Rescaled Image')
plt.xlabel('Gray Level Values')
plt.ylabel('Number of Pixels')

plt.tight_layout()
plt.show()
```

```python
#Task2
# Apply histogram equalization to the moon image 
img_eq = exposure.equalize_hist(img)

fig = plt.figure(figsize=(10, 8))

fig.add_subplot(2, 1, 1)
plt.imshow(img_eq, cmap='gray')
plt.title('Moon Image - Histogram Equalization')
plt.axis('off')

fig.add_subplot(2, 1, 2)
plt.hist(img_eq.flat, bins=256)
plt.title('Flattened Histogram')
plt.xlabel('Normalized Intensity')
plt.ylabel('Number of Pixels')

plt.tight_layout()
plt.show()
```

```python
#Task3
# Load the source and reference images 
source = data.chelsea()      # The cat image
reference = data.rocket()    # The rocket image (reference)

# Implement histogram matching [cite: 80, 84]
# channel_axis=-1 is used for multi-channel (RGB) images
matched = exposure.match_histograms(source, reference, channel_axis=-1)

# Display Source, Reference, and Matched images for comparison 
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

ax1.imshow(source)
ax1.set_title('Source (Chelsea)')
ax1.axis('off')

ax2.imshow(reference)
ax2.set_title('Reference (Rocket)')
ax2.axis('off')

ax3.imshow(matched)
ax3.set_title('Matched Result')
ax3.axis('off')

plt.tight_layout()
plt.show()
```
