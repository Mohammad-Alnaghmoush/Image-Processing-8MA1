```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('cameraman.tif')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
height, width = img.shape[:2]

scaled_img = cv2.resize(img, (int(width*1.5), int(height*1.5)))

plt.imshow(scaled_img)
plt.title("Scaled Image")
plt.axis("off")
plt.show()
```

```python
(h, w) = img.shape[:2]
center = (w // 2, h // 2)

rotation_matrix = cv2.getRotationMatrix2D(center, 120, 1.0)
rotated_img = cv2.warpAffine(img, rotation_matrix, (w, h))

plt.imshow(rotated_img)
plt.title("Rotated 120 Degrees")
plt.axis("off")
plt.show()
```

```python
rows, cols = img.shape[:2]

shear_matrix = np.float32([[1, 0.5, 0],
                           [0, 1, 0]])

sheared_img = cv2.warpAffine(img, shear_matrix, (cols, rows))

plt.imshow(sheared_img)
plt.title("Sheared Image")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
negative = 255 - gray

plt.imshow(negative, cmap='gray')
plt.title("Negative Image")
plt.axis("off")
plt.show()
```

```python
gray_float = gray.astype(np.float32)

c = 255 / np.log(1 + np.max(gray_float))

log_image = c * np.log(1 + gray_float)

log_image = np.clip(log_image, 0, 255)  # prevent overflow
log_image = log_image.astype(np.uint8)

plt.imshow(log_image, cmap='gray')
plt.title("Log Transformation")
plt.axis("off")
plt.show()
```

```python
gamma = 0.5
power_law = np.array(255 * (gray / 255) ** gamma, dtype='uint8')

plt.imshow(power_law, cmap='gray')
plt.title("Power Law (Gamma=0.5)")
plt.axis("off")
plt.show()
```
